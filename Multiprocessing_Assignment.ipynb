Q1. What is multiprocessing in python? Why is it useful?

* Multiprocessing in python is  a built-in package that allows the system to run multiple processes simultaneously.

* i.e. useful because provides a way to create and manage multiple processes. Each process runs in its own separate    memory space, allowing them to work independently of each other. 

Q2. What are the differences between multiprocessing and multithreading?

The main difference between these two is :

* In multithreading, multiple threads at the same time are generated by a single process. 
* In multiprocessing, multiple threads at the same time run across multiple cores.

Q3. Write a python code to create a process using the multiprocessing module.

```python
import multiprocessing

def test():
    print("Hii! This is Sandeep")
    
if __name__ == '__main__':
    m= multiprocessing.Process(target=test)
    
    m.start()
    m.join()
    
    print("Doneeee")
```

Q4. What is a multiprocessing pool in python? Why is it used?

Ans:  
*  Multiprocessing pool in python creates multiple processes in the background and spreads out computations across     multiple CPU cores so that they all happen in parallel without much human efforts.
*  We use Multiprocessing pool in python for parallel execution of a function across multiple input values,            distributing the input data across processes

Q5. How can we create a pool of worker processes in python using the multiprocessing module?

* We can create a pool of worker processes in Python using the multiprocessing module's  "Pool" class.
* The Pool class provides an easy way to manage and distribute tasks among multiple processes for parallel execution.

```python
#example :

import multiprocessing

def print_square(n):
    return n**2
if __name__ == '__main__':
    with multiprocessing.Pool(processes =4) as pool:
        out= pool.map(print_square , [1,2,3,4,5,6,7,8,9])
print(out)
```

Q6. Write a python program to create 4 processes, each process should print a different number using the
multiprocessing module in python.

```python
import multiprocessing

def print_number(number):
    print(f"Process {number}: Printing {number}")

if __name__ == '__main__':
    processes = []
    
    for i in range(1,5):
        process = multiprocessing.Process(target=print_number, args=(i,))
        processes.append(process)
        process.start()
        
    for process in processes :
        process.join()
        
    print("Finished....")
        
```
