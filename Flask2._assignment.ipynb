Q1. Explain GET and POST methods.

Ans:

 #GET Method:
* The GET method is used in HTTP to request data from a web server. It appends data to the URL as query parameters and is primarily used for retrieving information. It is a safe and idempotent operation.

 #POST Method:
* The POST method is used in HTTP to send data to a web server for processing. It doesn't append data to the URL but includes it in the request body. POST requests are typically used for creating or updating resources and can have side effects.

Q2. Why is request used in Flask?

* In Flask, the "request" object is used to access and handle incoming HTTP requests, allowing developers to extract data, parameters, and content sent by clients to the server.

Q3. Why is redirect() used in Flask?

* The redirect() function in Flask is used to perform a client-side redirection to a different URL. It's often used after processing a request to direct the user's browser to another page or URL, making it a useful tool for controlling navigation in web applications.

Q4. What are templates in Flask? Why is the render_template() function used?

* Templates are files that contain static data as well as placeholders for dynamic data. A template is rendered with specific data to produce a final document.

* The render_template() function in Flask is used to render and display HTML templates in a web application, dynamically populating them with data to create dynamic and customized web pages.
