Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class in Python is a base class for exceptions that occur when a specified key or index cannot be found or accessed within a sequence or dictionaries. 

```python
#KeyError:

my_dict = {'a': 1, 'b': 2, 'c': 3}

try:
    value = my_dict['d']
except KeyError as e:
    print("Error:", e)
```

```python
#IndexError:

my_list = [10, 20, 30]

try:
    value = my_list[3] 
except IndexError as e:
    print("Error:", e)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

Ans:

* ImportError and ModuleNotFoundError are both exceptions related to importing modules in Python.

* ImportError is a base class for exceptions that occur when there's an issue while importing a module or a symbol (function, class, etc.) from a module.

```python
try:
    import non_existent_module
except ImportError as e:
    print("Import Error:", e)
```

* ModuleNotFoundError: i.e. raised when a module cannot be found or imported. It's a subclass of ImportError and provides a clearer error message indicating that the module couldn't be located.

```python
try:
    import non_existent_module
except ModuleNotFoundError as e:
    print("Module Not Found Error:", e)
```
